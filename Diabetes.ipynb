**Diabetes Dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df=pd.read_csv('/content/diabetes.csv')
```

```python
df
```

```python
# Display the first few rows
print(df.head())
```

```python
# Create a NumPy array from a DataFrame column
age_array = df['Age'].values
insulin_array = df['Insulin'].values

print(age_array[:7])  # Display first 7 elements
print(insulin_array[:7])  # Display first 7 elements
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display summary statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Compute mean, median, and standard deviation
mean_values = df.mean()
median_values = df.median()
std_dev_values = df.std()

print("Mean Values:\n", mean_values,"\n")
print("Median Values:\n", median_values,"\n")
print("Standard Deviation Values:\n", std_dev_values)
```

```python
#Filters only Diabetic Persons
diabetic_df = df[df['Outcome'] == 1]
print(diabetic_df)
```

```python
# Histogram of a specific column, e.g., 'Glucose'
plt.hist(df['Glucose'], bins=30, edgecolor='black')
plt.title('Glucose Levels')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```

```python
x=df['Age']
```

```python
y=df['BMI']
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Scatter plotting
plt.scatter(x,y,c='pink')
```

```python
#Bar chart
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
categories = ['BMI', 'Glucose', 'Insulin', 'Age']

ax.bar(categories, np.random.rand(len(categories)))
```

```python
#Count the number of diabetic and non-diabetic persons.
print(df['Outcome'].value_counts())
```

```python
# Group by Outcome and calculate mean for each group
print(df.groupby('Outcome').mean())
```

```python
# Filter data: persons with Glucose level greater than 140
high_glucose = df[df['Glucose'] > 140]
print("\nData of persons with Glucose level greater than 140:")
print(high_glucose.head())
```

```python
# Select specific columns: Glucose, Insulin, BMI
selected_columns = df[['Glucose', 'Insulin', 'BMI']]
print(selected_columns.head())
```

```python
# Create a new column: Glucose_BMI
df['Glucose_BMI'] = df['Glucose'] / df['BMI']
print(df.head())
```

```python
# Pie chart of Outcome counts
outcome_counts = df['Outcome'].value_counts()
plt.pie(outcome_counts, labels=['Non-Diabetic', 'Diabetic'], autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Diabetic vs Non-Diabetic Persons')
plt.show()
```

```python
# Box plot of Glucose levels
plt.boxplot(df['Age'])
plt.title('Box Plot of Glucose Levels')
plt.ylabel('Glucose')
plt.show()
```

```python
# Line plot
plt.plot(df['BloodPressure'])
plt.title('BloodPressure Levels')
plt.xlabel('Index')
plt.ylabel('BloodPressure')
plt.show()
```
